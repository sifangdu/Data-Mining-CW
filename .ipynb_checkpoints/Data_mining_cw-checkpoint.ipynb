In EDUCATION column, '5' and '6' both encodes for 'unknown', and '0' is not defined in the data dictionary.

1.remove all 0s, 5s and 6s, and replace with Nan
2.covert 0, 5, 6 to 0 which encodes for 'unknown'

In MARRIAGE column, '0' is not defined in the dictionary.

Solutions: 
1.remove all 0s and replace with Nan
2.covert 0 to 3 which encodes for 'unknown'

```python
# 2.here I chose to convert 0,5,6 to 4 because the label 'others' does not contain more information than 'unknown',
# although it may have effect on the distribution of data and therefore affect the model performance.
train["EDUCATION"].replace([0,5,6], 4, inplace = True)
test["EDUCATION"].replace([0,5,6], 4, inplace = True)

train["MARRIAGE"].replace([0], 3, inplace = True)
test["MARRIAGE"].replace([0], 3, inplace = True)
```
